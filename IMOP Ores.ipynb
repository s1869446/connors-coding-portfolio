```python
#Import relevant packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
#Load main data set
data=pd.read_excel("IMOP Ores Practical.xlsx",header=3,skiprows=[15,16,17,18,24,25,26,27,28,29,30],usecols='A:L')
```

```python
#Effects on zincite saturation index
plt.figure(figsize=(15,7.5))
for i in data['Cl-'].unique():
    cl=data[data['Cl-']==i]
    plt.plot(cl['T'],cl['SI Zincite'],label=str(i)+'M[Cl$^-$]')
    plt.scatter(cl['T'],cl['SI Zincite'],marker='s')
plt.ylabel('Zincite Saturation Index',fontsize=20)
plt.xlabel('Temperature ($^\circ$C)',fontsize=20)
plt.legend(fontsize=20)
```

```python
#Effects on Zn speciation
index=1
plt.figure(figsize=(25,25))
for j in data['T'].unique():
    T=data[data['T']==j]
    plt.subplot(2,2,index)
    for i in data.columns[6:]:
        plt.plot(T['Cl-'],np.log10(T[i]),label=i)
        plt.scatter(T['Cl-'],np.log10(T[i]),marker='s')
    plt.legend(loc=8,fontsize=20)
    plt.xlabel('[Cl$^-$] (mol/L)',fontsize=20)
    plt.ylabel('log(activity)',fontsize=20)
    plt.ylim(-9,-2)
    plt.title(str(j)+'$^\circ$C',fontsize=20)
    index+=1
```

```python
#Load pH data
ph=pd.read_excel("IMOP Ores Practical.xlsx",sheet_name='pH')
```

```python
#Effects of pH
plt.figure(figsize=(15,7.5))
plt.subplot(121)
plt.plot(ph['pH'],ph['SI Zincite'],c='r')
plt.scatter(ph['pH'],ph['SI Zincite'],c='r',marker='s')
plt.xlabel('pH',fontsize=20)
plt.xlim(6,1.5)
plt.ylabel('Zincite Saturation Index',fontsize=20)
plt.subplot(122)
for i in ['Zn2+', 'ZnOH+', 'ZnCl+']:
    plt.plot(ph['pH'],np.log10(ph[i]),label=i)
    plt.scatter(ph['pH'],np.log10(ph[i]),marker='s')
plt.legend(fontsize=20)
plt.xlabel('pH',fontsize=20)
plt.xlim(6,1.5)
plt.ylabel('log(activity)',fontsize=20)
```

```python
#Load sulfur data
sulf=pd.read_excel("IMOP Ores Practical.xlsx",sheet_name='Sulfur')
```

```python
#Effects of sulfur
plt.figure(figsize=(15,7))
plt.plot(sulf['HS-'],sulf['SI Zincite'],label='Zincite')
plt.scatter(sulf['HS-'],sulf['SI Zincite'])
plt.plot(sulf['HS-'],sulf['SI Sphalerite'],label='Sphalerite')
plt.scatter(sulf['HS-'],sulf['SI Sphalerite'])
plt.xscale('log')
plt.xlabel('[HS$^-$] (mol/L)',fontsize=20)
plt.ylabel('Saturation Index',fontsize=20)
plt.legend(fontsize=20)
```
